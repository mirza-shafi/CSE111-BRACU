```python
#task-01
class Student:
    ID = 0

    def __init__(self,name,dep,age,cgpa):
        self.name = name
        self.dept = dep
        self.age = age
        self.cgpa = cgpa
        Student.ID += 1

    def get_details(self):
        print(f'ID: {Student.ID}\nName: {self.name}\nDepartment: {self.dept}\nAge: {self.age}\nCGPA: {self.cgpa}')

    @classmethod
    def from_String(cls, string):
        name,dep,age,cgpa = string.split('-')
        n_student = cls(name,dep,age,cgpa)
        return n_student

s1 = Student("Samin", "CSE", 21, 3.91) 
s1.get_details() 
print("-----------------------") 
s2 = Student("Fahim", "ECE", 21, 3.85) 
s2.get_details() 
print("-----------------------") 
s3 = Student("Tahura", "EEE", 22, 3.01) 
s3.get_details()  
print("-----------------------") 
s4 = Student.from_String("Sumaiya-BBA-23-3.96") 
s4.get_details()

#subtask 5
#print('\nThe difference between a class variable & an instance variable is that a class variable can be accessed by all the objects of that same class but an instance variable can be accessed only by a specific created object of a class and cannot be accessed by all other objects of that class.')

#subtask 6
#print("\nThe difference between an instance method & a class method is that an instance method needs an instance of a class and takes 'self' as a parameter to access through that instance object but a class method does not need a class instance and takes 'cls' as a parameter to have access throughout the class itself.")
```

```python
#task-02
class Assassin:
    count = 0 

    def __init__(self,name,s_rate):
        if s_rate<=100:
            self.name = name
            self.success = s_rate
            Assassin.count+=1

    def printDetails(self):
        print(f'Name: {self.name}\nSuccess Rate: {self.success}%\nTotal number of Assassin: {Assassin.count}')

    @classmethod
    def failureRate(cls, name, fail):
        success = 100 - fail
        mod = cls(name,success)
        return mod

    @classmethod
    def failurePercentage(cls, name, fp):
        fail = int(fp[:-1])
        success = 100 - fail
        mod = cls(name,success)
        return mod

john_wick = Assassin('John Wick', 100) 
john_wick.printDetails() 
print('================================') 
nagisa = Assassin.failureRate("Nagisa", 20) 
nagisa.printDetails() 
print('================================') 
akabane = Assassin.failurePercentage("Akabane", "10%") 
akabane.printDetails()
```

```python
#task-03
class Passenger:
    count = 0

    def __init__(self,name):
        self.name = name
        self.bag = 0
        self.fare = 450
        Passenger.count += 1

    def set_bag_weight(self,bag):
        self.bag = bag
        if self.bag<=20:
            self.fare += 0
        elif self.bag<=50:
            self.fare += 50
        else:
            self.fare += 100

    def printDetail(self):
        print(f'Name: {self.name}\nBus Fare: {self.fare} taka')

print("Total Passenger:", Passenger.count) 
p1 = Passenger("Jack") 
p1.set_bag_weight(90) 
p2 = Passenger("Carol") 
p2.set_bag_weight(10) 
p3 = Passenger("Mike") 
p3.set_bag_weight(25) 
print("=========================") 
p1.printDetail() 
print("=========================") 
p2.printDetail() 
print("=========================") 
p3.printDetail() 
print("=========================") 
print("Total Passenger:", Passenger.count) 
```

```python
#Task 5
class Employee:
    def __init__(self,name,exp):
        self.name = name
        self.workingPeriod = exp

    @classmethod
    def employeeByJoiningYear(cls,name,year):
        exp = 2022 - year
        mod = cls(name,exp)
        return mod

    @staticmethod
    def experienceCheck(year,gender):
        if year<3:
            if gender == 'male':
                return 'He is not experienced'
            elif gender == 'female':
                return 'She is not experienced'

        else:
            if gender == 'male':
                return 'He is experienced'
            elif gender == 'female':
                return 'She is experienced'

employee1 = Employee('Dororo', 3) 
employee2 = Employee.employeeByJoiningYear('Harry', 2016) 
print(employee1.workingPeriod) 
print(employee2.workingPeriod) 
print(employee1.name) 
print(employee2.name) 
print(Employee.experienceCheck(2, "male")) 
print(Employee.experienceCheck(3, "female"))
```

```python
#Task 6
class Laptop:
    laptopCount = 0

    def __init__(self,name,count):
        self.name = name
        self.count = count
        Laptop.laptopCount += self.count

    @staticmethod
    def advantage():
        print('Laptops are portable')

    @classmethod
    def resetCount(cls):
        Laptop.laptopCount = 0

lenovo = Laptop("Lenovo", 5); 
dell = Laptop("Dell", 7); 
print(lenovo.name, lenovo.count) 
print(dell.name, dell.count) 
print("Total number of Laptops", Laptop.laptopCount) 
Laptop.advantage() 
Laptop.resetCount() 
print("Total number of Laptops", Laptop.laptopCount)
```

```python
#Task 7

class Cat:
    Number_of_cats = 0

    def __init__(self,color,action):
        self.color = color
        self.action = action
        Cat.Number_of_cats += 1

    @classmethod
    def no_parameter(cls):
        name = 'White'
        action = 'sitting'
        mod_no = cls(name,action)
        return mod_no

    @classmethod
    def first_parameter(cls,color):
        action = 'sitting'
        mod_one = cls(color,action)
        return mod_one

    @classmethod
    def second_parameter(cls,action):
        color = 'Grey'
        mod_two = cls(color,action)
        return mod_two

    def printCat(self):
        print(f'{self.color} cat is {self.action}')

    def changeColor(self,color):
        self.color = color

print("Total number of cats:", Cat.Number_of_cats) 
c1 = Cat.no_parameter() 
c2 = Cat.first_parameter("Black") 
c3 = Cat("Brown", "jumping") 
c4 = Cat("Red", "purring") 
c5 = Cat.second_parameter("playing") 
print("=======================") 
c1.printCat() 
c2.printCat() 
c3.printCat() 
c4.printCat() 
c5.printCat() 
c1.changeColor("Blue") 
c3.changeColor("Purple") 
c1.printCat() 
c3.printCat() 
print("=======================") 
print("Total number of cats:", Cat.Number_of_cats) 
```

```python
#Task 8

import math
class Cylinder:
    radius = 5
    height = 18

    def __init__(self,r,h):
        self.radius = r
        self.height = h
        print(f'Default radius={Cylinder.radius} and height={Cylinder.height}.')
        Cylinder.radius = self.radius
        Cylinder.height = self.height
        print(f'Updated: radius={Cylinder.radius} and height={Cylinder.height}.')

    @classmethod
    def swap(cls,r,h):
        swapped = cls(h,r)
        return swapped

    @classmethod
    def changeFormat(cls,info):
        r,h = info.split('-')
        mod = cls(float(r),float(h))
        return mod

    @staticmethod
    def area(r,h):
        area = 2*math.pi*r**2 + 2*math.pi*r*h
        print('Area:',area)

    @staticmethod
    def volume(r,h):
        volume = math.pi*r**2*h
        print('Volume:',volume)

c1 = Cylinder(0,0) 
Cylinder.area(c1.radius,c1.height) 
Cylinder.volume(c1.radius,c1.height) 
print("===============================") 
c2 = Cylinder.swap(8,3) 
c2.area(c2.radius,c2.height) 
c2.volume(c2.radius,c2.height) 
print("===============================") 
c3 = Cylinder.changeFormat("7-13") 
c3.area(c3.radius,c3.height) 
c3.volume(c3.radius,c3.height) 
print("===============================") 
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height) 
print("===============================") 
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```
